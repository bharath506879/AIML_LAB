```python
!pip install pandas
import pandas as pd
import matplotlib.pyplot as plt
people=['amulya','ramya','siri','sravya']
age=[12,43,16,17]
height=[5.5,5.3,4.9,5.5]
weight=[33,35,45,37]
data ={
    'name':people,
    'Age':age,
    'Height':height,
    'Weight':weight,
}
df=pd.DataFrame(data)
plt.figure(figsize=(8,6))
plt.scatter(df['Age'],df['Weight'],color='blue',marker='o')
plt.title('Scatter plot of Age distribution')
plt.xlabel('Age(years)')
plt.ylabel('Weight(kg)')
plt.grid(True)
plt.show()
```
